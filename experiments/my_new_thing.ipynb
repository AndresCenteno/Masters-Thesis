```python
from scipy.linalg import expm, logm
import numpy as np
import pygsp as pg
```

```python
def create_signals(L,dt,x0):
    n = L.shape[0]
    X = np.zeros([n,n+1])
    X[:,0] = x0
    expL = expm(-L*dt)
    for i in range(1,n+1):
        X[:,i] = expL@X[:,i-1]
    return X
```

```python
def solve_for_signal(X,dt):
    n = X.shape[0]
    signals = X[:,0:n]
    shifted_signals = X[:,1:n+1]
    print(shifted_signals.shape[:])
    return -logm(shifted_signals@np.linalg.inv(signals))/dt
```

```python
N = 20
dt = 0.5
L = lh.heat_graph_RBF(N)
x0 = np.random.rand(N)
X = create_signals(L,dt,x0)
X[:,0:-1].shape[:]
L_learned = solve_for_signal(X,dt)
L_learned[abs(L_learned)<0.02] = 0
L_learned = np.real(L_learned)
L_learned = N*L_learned/np.trace(L_learned)
np.linalg.norm(L_learned-L)
```
