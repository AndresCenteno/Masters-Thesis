```python
import sys
sys.path.insert(1,'../..')
import learnHeat as lh
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = lh.heat_graph_BA(40)
v = lh.laplacian_to_vec(G)
plt.hist(v[v!=0],bins=20,range=[0,max(v)])
plt.xlabel("Weight bins")
plt.ylabel("Number of edges in each weight bin")
```

```python
tau = [0.5,1,2.5,4]
X = lh.create_deltas(G,tau)
res = lh.learn_heat(X,tau0=tau,alpha=0.01,beta=0.1,max_iter=50)
L = res["L"]
```

```python
vec=lh.laplacian_to_vec(L)
n, bins, _ = plt.hist(vec,bins=5)
plt.axvspan(bins[7], bins[10], facecolor='yellow', alpha=0.5,label="Plateau")
plt.axvspan(bins[6], bins[7], facecolor='purple', alpha=0.5,label="Threshold")
plt.axvline(0.18,color='green',linestyle='--',label='Median bin')
plt.legend()
plt.ylim(0, 100)  # Adjust the values based on your data
plt.text(0.17, 90, 'No changes\nin edges', color='red')
plt.xlabel("Weight bins")
plt.ylabel("Number of edges per bin")
plt.title(r"BA graph, $N=40$, partition size $=16$")
```

```python
vec=lh.laplacian_to_vec(L)
hist, bins, content = plt.hist(vec,bins=200)
empty_bins = []
for i in range(50, len(bins)-1):
    if hist[i] ==0 and hist[i-1] > 0 and hist[i+1] > 0:
        empty_bins.append(i)

for bin_idx in empty_bins:
    plt.axvline(bins[bin_idx], color='yellow')
plt.axvline(bins[empty_bins[0]], color='yellow',label='Least persistent plateaus')
plt.axvline(bins[100],linestyle='--',color='green',label=r"$w_{max}/2$")
plt.ylim(0, 10)  # Adjust the values based on your data
plt.xlabel("Weight bins")
plt.ylabel("Number of edges per bin")
plt.legend()
plt.title(r"BA graph, $N=40$, partition size $=200$")
```
